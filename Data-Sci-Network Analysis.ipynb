# Simulating a social network: directed network analysis steps:
### generate graph object and edge list
### assign attributes to graph nodes
### visualize the network

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import networkx as nx

%matplotlib inline 
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

## Generating a graph object and edgelist

```python
DG = nx.gn_graph(7, seed = 25)

#creating an edge list
for line in nx.generate_edgelist(DG, data = False): print(line)
```

### the function generated and edge from node 1 to 0, 2 to 0, 3 to 2 and so on

## Assigning attributes to nodes
### we can attach attributes to nodes or edges
### attachable attributes include: weight, labels, colors and other python objects


```python
print (DG.node[0])#nothing in the dictionary 
```

```python
DG.node[0]['name'] = 'Alice'
```

```python
print (DG.node[0])
```

```python
DG.node[1]['name'] = 'Bob'
DG.node[2]['name'] = 'Claire'
DG.node[3]['name'] = 'Dennis'
DG.node[4]['name'] = 'Esther'
DG.node[5]['name'] = 'Frank'
DG.node[6]['name'] = 'George'
```

```python
DG.add_nodes_from([(0,{'age':25}),(1,{'age':31}),(2,{'age':18}),(3,{'age':47}),(4,{'age':22}),
                   (5,{'age':23}),(6,{'age':50})])
print (DG.node[0])
```

```python
# adding a gender attribute and adding a gender
DG.node[0]['gender'] = 'f'
DG.node[1]['gender'] = 'm'
DG.node[2]['gender'] = 'f'
DG.node[3]['gender'] = 'm'
DG.node[4]['gender'] = 'f'
DG.node[5]['gender'] = 'm'
DG.node[6]['gender'] = 'm'
```

## Visualize the network graph

```python
nx.draw_circular(DG, node_color='bisque', with_labels=True)
```

### In this case node 1, 2 and 5 follow node 0, node 0 does not follow back

```python
#passing a dictionary to label graph and pass key value pairs of node:name
labeldict = {0: 'Alice',1:'Bob',2:'Claire',3:'Dennis',4:'Esther',5:'Frank',6:'George'}

nx.draw_circular(DG, labels=labeldict, node_color='bisque', with_labels=True)
```

```python
#transfroming a directed graph to undirected
G = DG.to_undirected()
```

```python
nx.draw_spectral(G, labels=labeldict, node_color='bisque', with_labels=True)
```

# Social Network Analysis metrics 
### degree describes a node's connectedness
### Sucessors: a sucessor node is a node that could serve as a backup and potentially replace an influencial node in a network. The function of a sucessor is to preserve the flow of influence throughout a network in the case where an important node is removed 
### Neighbors: adjacent nodes in a network
### In-degree: the theory is that if a person or a profile has a lot of inbound connections then they're prestigious because many different people and profiles want to connect with them
### Out-degree: if a person or profile has a lot of outgoing conections, however, then they're still considered influential. That's because in theory these people have more of a platform accross which to engage and interact with their many outbound connections


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import networkx as nx

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

```python
DG = nx.gn_graph(7, seed = 25)

for line in nx.generate_edgelist(DG, data=False):
    print(line)

DG.node[0]['name'] = 'Alice'
DG.node[1]['name'] = 'Bob'
DG.node[2]['name'] = 'Claire'
DG.node[3]['name'] = 'Dennis'
DG.node[4]['name'] = 'Esther'
DG.node[5]['name'] = 'Frank'
DG.node[6]['name'] = 'George'
```

```python
G = DG.to_undirected()
```

```python
print (nx.info(DG))
```

### The average degree of incoming connections in this particular graph is .85, the same as the outgoing
### but alice has just ingoing edges, she might be the most influential node

```python
DG.degree()
```

### checking the degrees confirms that alice has 3 connections 
### but, let's check sucessor nodes

```python
#identifying sucessor nodes (could replace and maintain influence)
nx.draw_circular(DG, node_color='bisque', with_labels=True)
```

```python
list(DG.successors(3))
```

### Python predicts that node 2 is a successor for node 3 

### Looking at the neighbors function: works differently in directed graphs vs undirected graphs, it returns a list of all outbound connections so for example for node 4 it would only return node 3 but not node 6 since node 6 is inbound, not outbound

```python
list(DG.neighbors(4))
```

### it we call neighbors on an undirected graph it will return all the connections

```python
list(G.neighbors(4))
```
