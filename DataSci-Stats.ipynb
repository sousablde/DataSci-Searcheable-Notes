# Parametric correlation analysis


## Pearson Correlation Coefficient
### R = 1 - strong positive relatioship
### R = 0 - not linearly correlated
### R = -1 - strong negative correlation

## Assumptions: data normally distributed, continuous numeric variables, variables are linearly related

### Used to uncover linear relatioships between variables
### Not to rule out possible nonlinear relatioships between variables


```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sb

from pylab import rcParams
from pandas import Series, DataFrame
from scipy import stats
from scipy.stats import pearsonr
```

## Using scipy to calculate pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coefficient, p_value = pearsonr(mpg, hp)

#printing a label for the results of this r test
print ('PearsonR Correlation Coefficient ', pearsonr_coefficient)
```

### The neagtive value of R means that the two varialbles are negatively correlated you will get less miles per gallon on a car that has greater horse power

```python
pearsonr_coefficient, p_value = pearsonr(mpg, qsec)

#printing a label for the results of this r test
print ('PearsonR Correlation Coefficient for the pair mpg and qsec is ', pearsonr_coefficient)
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, wt)

#printing a label for the results of this r test
print ('PearsonR Correlation Coefficient for the pair mpg and wt is ', pearsonr_coefficient)
```

### the strongest linear correlation was between weight and mpg, the heaviest the car the lower will be the number of miles achieved per gallon

# Non Parametric methods for data analysis
## correlation for categorical, non normal, non linearly related data
## ex smoking habits and income 

## Spearman's rank correlation: finds r for variable pairs of ordinal (numerical value that can be categorized) data types, variable pairs are then able to be ranked according to the strength fo the correlation between them 

### R = 1 - strong positive relatioship
### R = 0 - not linearly correlated
### R = -1 - strong negative correlation

### Spearman's assumptions: variables are ordinal (numeric but can be ranked like a categorical variable), model assumes variables are related non-linearly and are not normally distributed

## Chi-square tables test for independence
### p<0.005 - we reject the null hypothesis and conclude the variables are correlated
### p>0.05 - accept null hypothesis and conclude the variables are independent

### use it when variables are categorical or numeric 
### there are binned numeric variables (example when a variable can be grouped in for instance 10 groups fo 10)

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sb

from pylab import rcParams
from pandas import Series, DataFrame
from scipy import stats
from scipy.stats import spearmanr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 14, 7
plt.style.use('seaborn-whitegrid')
```

## Spearman Rank Correlation

```python
address = r'C:\Users\sousa\Downloads\Python for Data Science Essential Training\Ex_Files_Python_Data_Science_EssT\Ex_Files_Python_Data_Science_EssT\Exercise Files\Ch01\01_05\mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
#making a scatter plot to check how the variables relate to each other

sb.pairplot(cars)
```

```python
#we will select a few that are good candidates because they are ordinal and are not linear
X = cars[['cyl', 'vs', 'am', 'gear']]
sb.pairplot(X)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl, vs)

#print ('Spearman Rank Correlation is ', spearmanr_coefficient)
print ('Spearman Rank Correlation is {0:.2f}'.format(spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, am)
print ('Spearman Rank Correlation is {0:.2f}'.format(spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, gear)
print ('Spearman Rank Correlation is {0:.2f}'.format(spearmanr_coefficient))
```

### the strongest correlation is found on the pair cyl/vs

## Chi-square test for independence
### variables need to be in a crosstab table


```python
from scipy.stats import chi2_contingency
table = pd.crosstab(cyl, am)

chi2, p, dof, expected = chi2_contingency(table.values)

print ('Chi-square Stats are {0:.2f}'.format(chi2))
print ('Chi-square Stats p_value {0:.3f}'.format(p))
```

```python
table = pd.crosstab(cyl, vs)

chi2, p, dof, expected = chi2_contingency(table.values)

print ('Chi-square Stats are {0:.2f}'.format(chi2))
print ('Chi-square Stats p_value {0:.3f}'.format(p))
```

```python
table = pd.crosstab(cyl, gear)

chi2, p, dof, expected = chi2_contingency(table.values)

print ('Chi-square Stats are {0:.2f}'.format(chi2))
print ('Chi-square Stats p_value {0:.3f}'.format(p))
```

### for variables to be independent p value needs to be greater than 0.05, because none of these p values abide by that we reject the null hypothesis and conclude that the variables are in fact correlated

## Normalizing and transforming features with MinMaxScalar() and fit_transform()

```python
address = r'C:\Users\sousa\Downloads\Python for Data Science Essential Training\Ex_Files_Python_Data_Science_EssT\Ex_Files_Python_Data_Science_EssT\Exercise Files\Ch01\01_05\mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
mpg = cars.mpg
plt.plot(mpg)
```

#### max and min values range between 10 and 35 the average is about 20

```python
cars[['mpg']].describe()
```

```python
#place value in one column matrix
mpg_matrix = mpg.values.reshape(-1,1)

scaled = preprocessing.MinMaxScaler()

scaled_mpg = scaled.fit_transform(mpg_matrix)

plt.plot(scaled_mpg)
```

```python
#defining a manual feature for the range
mpg_matrix = mpg.values.reshape(-1,1)

scaled = preprocessing.MinMaxScaler(feature_range = (0,10))

scaled_mpg = scaled.fit_transform(mpg_matrix)

plt.plot(scaled_mpg)
```

### scaling features and standardizing variables

```python
standardized_mpg = scale(mpg, axis = 0, with_mean = False, with_std = False)
plt.plot(standardized_mpg)
```

### we get back our scaled untransformed data but with the scale function it just carries the default 

```python
standardized_mpg = scale(mpg)
plt.plot(standardized_mpg)
```

### the mean is centered around 0 in this case and the distribution has now unit variance (standard deviation) it is now standardized
